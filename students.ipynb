1: GRAPH SHOWING THE PARENTAL LEVEL OF EDUCATION AND SCORE IN THE MATH TEST FOR THE MALE STUDENTS IN THE GROUP A AT SOME COLLEGE WITH STANDARD LUNCH AND FREE/REDUCED.


```python
pl = students[students['gender'] == 'male'] 
pa = pl[pl['race/ethnicity'] == 'group A'] 
pa
```

5: GRAPH TO SHOW THE HIGHEST AND LOWEST SCORES IN THE READING TEST IN DESCENDING ORDER

```python
fig = plt.figure(figsize = (20,15))
df_sorted_desc= students.sort_values('reading score',ascending=False)
plt.plot(students['reading score'], data = df_sorted_desc,color = 'b', ls = '--', lw = 0.2, marker = 'o',
         markerfacecolor = 'k', markeredgecolor = 'r')
plt.title('HIGHEST TO LOWEST SCORE OF READING TEST IN DESCENDING ORDER', color = 'black', fontsize = 20)
plt.show()
```

The graph shows that the highest score in the reading test was 100, while the lowest was below 20

7: GRAPH SHOWING THE HIGHEST AND LOWEST SCORES IN THE WRITING TEST

```python
fig = plt.figure(figsize = (20,15))
df_sorted_desc= students.sort_values('writing score',ascending=False)
plt.plot(students['writing score'], data = df_sorted_desc,color = 'b', ls = '--', lw = 0.2, marker = 'h',
         markerfacecolor = 'r', markeredgecolor = 'black')
plt.title('HIGHEST TO LOWEST SCORE OF WRITING TEST IN DESCENDING ORDER', color = 'black', fontsize = 20)
plt.show()
```

The graph shows that the highest score obtained for the writing test is 100, while the lowest score is below 20

8: CONVERTING DATA FRAME TO EXCEL WITH EACH GROUP IN A DIFFERENT SHEET

```python
groupa = students[students['race/ethnicity'] == 'group A']
groupb = students[students['race/ethnicity'] == 'group B']
groupc = students[students['race/ethnicity'] == 'group C']
groupd = students[students['race/ethnicity'] == 'group D']
groupe = students[students['race/ethnicity'] == 'group E']

writer = pd.ExcelWriter('students2.xlsx')
groupa.to_excel(writer, 'GROUP A')
groupb.to_excel(writer, 'GROUP B')
groupc.to_excel(writer, 'GROUP C')
groupd.to_excel(writer, 'GROUP D')
groupe.to_excel(writer, 'GROUP E')
writer.save()
```
