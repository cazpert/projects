 2: Carry out a research on how question 1 can be represented with graph and plot it


```python
fig = plt.figure(figsize= (20,15))
genre = best['Genre'].value_counts().plot.pie(autopct = '%1.2f%%', labels = ['NON-FICTION:310', 'FICTION : 210']).set_title(
    'PERCENTAGE OF FICTION AND NON-FICTION BOOKS', color = 'black', fontsize = 20)
plt.show()
```

3: a graph showing the number of fiction and non-fiction book by year and state the highest and the lowest year

```python
fig = plt.figure(figsize = (20,15))
sns.set_palette('bright')
sns.countplot(x = best['Year'], hue = best['Genre']).set_title('NUMBER OF FICTION AND NON-FICTIONS BOOKS BY YEAR', 
                                                                 color = 'black', fontsize = 20)
plt.show()
```

The highest year fiction is 2014 while the highest year for non- fiction is 2015


The lowest year for fiction is 2015 while the lowest year for non fiction is 2014


4: a graph showing the distribution of price by genre

```python
fig = plt.figure(figsize = (20,20))
sns.set_palette('dark')
sns.violinplot(x = best['Genre'], y = best['Price'], data = best, 
               split = True).set_title('THE DISTIBUTION OF PRICE BY GENRE', color = 'black', fontsize = 20)
plt.show()
```

5: a graph showing the mean price of fiction and non-fiction books by year

```python
f = np.mean(best[best['Genre'] == 'Fiction']['Price'])
f_mean = np.mean(f)
print(f'The mean price for fiction is {f_mean}')
```

```python
fig = plt.figure(figsize = (20,20))
sns.kdeplot(best[best['Genre'] == 'Fiction']['Price'], color = 'g', fill = True)
plt.show()
```

```python
f = np.mean(best[best['Genre'] == 'Non Fiction']['Price'])
nf_mean = np.mean(f)
print(f'The mean price for Non fiction is {nf_mean}')
```

```python
fig = plt.figure(figsize = (10,10))
sns.kdeplot(best[best['Genre'] == 'Non Fiction']['Price'], color = 'purple', fill = True)
plt.show()
```

6: List out the 5 most expensive fiction and non-fiction books in dataframe with the following features(Authur,Name and price)

```python
ex_fiction = best[best['Genre'] == 'Fiction']
ex_fiction.drop(['User Rating', 'Reviews', 'Year'], axis = 1, inplace = True)
ex_fiction.nlargest(5, ['Price'])
```

```python
ex_nfiction = best[best['Genre'] == 'Non Fiction']
ex_nfiction.drop(['User Rating', 'Reviews', 'Year'], axis = 1, inplace = True)
ex_nfiction.nlargest(5, ['Price'])
```
